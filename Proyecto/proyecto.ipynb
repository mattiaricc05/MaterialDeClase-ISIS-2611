```python
from langdetect import detect, detect_langs, LangDetectException

def detect_language(text: str) -> str:
    try:
        return detect(text[:500])  # langdetect es rápido con fragmentos
    except LangDetectException:
        return "unknown"

def detect_language_probs(text: str) -> list:
    try:
        return detect_langs(text[:500])
    except LangDetectException:
        return []

df["lang"] = df["text_clean"].apply(detect_language)

print(df["lang"].value_counts())
# Para textos en español antiguo, espera mezcla de "es", "pt", "la", "ca"
```
